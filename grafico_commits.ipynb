```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

grafico = pd.read_csv("repo_commit_issues_data.csv")

display(grafico)

# Função para gerar gráficos separados para cada repositório
def plot_commit_issue_pull_per_repo(df):
    # Iterar sobre cada repositório no DataFrame
    for index, row in df.iterrows():
        repo_name = row['repo_name']
        
        # Criar subplots para cada repositório
        fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))
        fig.suptitle(f'Repositório: {repo_name}', fontsize=16)
        
        # Plotar os commits semanais, mensais e anuais
        axes[0].bar(['Weekly Commits', 'Monthly Commits', 'Yearly Commits'], 
                    [row['weekly_commits'], row['monthly_commits'], row['yearly_commits']], color='red')
        axes[0].set_title('Commits')
        axes[0].set_ylabel('Número de Commits')

        # Plotar as issues e pull requests
        axes[1].bar(['Open Issues', 'Closed Issues', 'Open Pulls', 'Closed Pulls'], 
                    [row['open_issues'], row['closed_issues'], row['open_pulls'], row['closed_pulls']], color='blue')
        axes[1].set_title('Issues e Pull Requests')
        axes[1].set_ylabel('Número de Issues/Pulls')

        # Ajustar layout e exibir gráfico
        plt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.show()

# Exemplo de uso com o DataFrame
plot_commit_issue_pull_per_repo(grafico)
```
